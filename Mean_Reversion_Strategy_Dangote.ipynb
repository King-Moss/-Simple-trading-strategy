```python
#Read the data
df = pd.read_csv('Dangote Cement Stock Price History (15) (1).csv')
#set the date as the index
df = df.set_index(pd.DatetimeIndex(df['Date'].values))
#Remove or drop the date column
df.drop(['Date'], axis=1, inplace=True)
# show the data
df
```

```python
#create a funtion to calculate the sma
def SMA(data, period=30, column='Price'):
  return data[column].rolling(window=period).mean()
```

```python
# Ensure 'Price' is numeric
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

# Define SMA function
def SMA(df, window):
    return df['Price'].rolling(window=window).mean()

# Add SMA to dataframe
df['SMA'] = SMA(df, 21)

# Calculate returns
df['Simple_Returns'] = df['Price'].pct_change(1)
df['Log_Returns'] = np.log(1 + df['Simple_Returns'])

# Calculate Price/SMA ratio
df['Ratios'] = df['Price'] / df['SMA']

df
```

```python
#Get some statistics on the ratio column
df['Ratios'].describe()
```

```python
#Get and show the percentile values
percentiles = [15, 20, 50, 80, 85]
#Remove any NA values in the Ratios Column and store the result in a new variable called ratios
ratios = df['Ratios'].dropna()
#Get the values of the percentiles
percentiles_values = np.percentile(ratios, percentiles)
#Show the values of the percentiles
percentiles_values
```
